## (utils) Json file 내 키워드 포함 조항 찾기

````python
import json

def print_clause(clauses: list):
    for clause in clauses:
        print(f'{clause["index"]} : {clause["content"]}')

def search_json(json_path, word1, word2):
    clauses_and = []
    clauses_1 = []
    clauses_2 = []
    clauses_total = []

    with open(json_path, 'r', encoding='utf-8') as f:
        data = json.load(f)

    for clause in data:
        if word1 in clause["content"] and word2 in clause["content"]:
            clauses_and.append(clause)
        if word1 in clause["content"]:
            clauses_1.append(clause)
            clauses_total.append(clause)
        if word2 in clause["content"]:
            clauses_2.append(clause)
            clauses_total.append(clause)
        
    
    print(f'## {word1} & {word2} ##')
    if len(clauses_and) > 0:
        print_clause(clauses_and)
    else:
        print("공통조항 없음")
    
    print(f'\n\n## {word1} 언급 조항 ##')
    if len(clauses_1) > 0:
        print_clause(clauses_1)
    else:
        print(f"{word1} 언급 조항 없음")
    
    print(f'/n/n## {word2} 언급 조항 ##')
    if len(clauses_2) > 0:
        print_clause(clauses_2)
    else:
        print(f"{word2} 언급 조항 없음")

    return clauses_and, clauses_1, clauses_2, clauses_total
````

````python
def split_text(text):
    result = []
    buffer = ''
    stack = []  # To keep track of parentheses
    i = 0
    while i < len(text):
        c = text[i]
        if c == '(':
            stack.append('(')
            buffer += c
            i += 1
        elif c == ')':
            if stack:
                stack.pop()
            buffer += c
            i += 1
        elif c == '/n':
            # Do not split if newline is immediately before a letter
            if i + 1 < len(text) and text[i + 1].isalpha():
                buffer += c
            else:
                if buffer.strip():
                    result.append(buffer.strip())
                buffer = ''
            i += 1
        elif c == '.':
            # Check if inside parentheses or after a number
            if stack or (i > 0 and text[i - 1].isdigit()):
                buffer += c
            else:
                buffer += c
                if buffer.strip():
                    result.append(buffer.strip())
                buffer = ''
            i += 1
        else:
            buffer += c
            i += 1
    if buffer.strip():
        result.append(buffer.strip())
    return result
````

````python
import copy

keyword_list = set()
keyword_node_list = []

def make_node(clause, keyword: str):

    global keyword_list, keyword_node_list

    if keyword not in keyword_list:
        node = copy.deepcopy(clause)
        node["keyword_name"] = keyword
        keyword_list.add(keyword)
        keyword_node_list.append(node)

    else:
        # Find the node corresponding to the keyword
        node = [x for x in keyword_node_list if x["keyword_name"] == keyword][0]
        keyword_node_list = [x for x in keyword_node_list if x["keyword_name"] != keyword]

        # Append values to the existing node
        node["index"] += clause["index"]
        node["subtitle"] += clause["subtitle"]
        node["metadata"]["title"]["doc"] += clause["metadata"]["title"]["doc"]
        node["metadata"]["title"]["chapter"] += clause["metadata"]["title"]["chapter"]
        node["metadata"]["title"]["section"] += clause["metadata"]["title"]["section"]
        node["metadata"]["title"]["subsection"] += clause["metadata"]["title"]["subsection"]

        keyword_node_list.append(node)

    return node
    
````

## 사용자 쿼리와의 임베딩 유사도로 top-k clause 뽑기


````python
from GraphDB.utils import text_embed

query1['embedding_text'] = text_embed(query1['text'], config['embedding_model'])
query2['embedding_text'] = text_embed(query2['text'], config['embedding_model'])
````

````python
def get_top_k_similar_nodes(embedding_text, k=3):
    with dbms.driver.session(database="clause-graph") as session:
        result = session.run("""
        MATCH (n)
        WHERE n.embedding_text IS NOT NULL
        WITH n, gds.similarity.cosine(n.embedding_text, $embedding_text) AS similarity
        RETURN n, similarity
        ORDER BY similarity DESC
        LIMIT $k
        """,
        embedding_text=embedding_text,
        k=k)
        
        nodes = []

        for record in result:
            node = record["n"]
            similarity = record["similarity"]
            nodes.append((node, similarity))

        return nodes
````

````python
top_nodes_1 = get_top_k_similar_nodes(query1['embedding_text'])
for node, similarity in top_nodes_1:
    print(f"Node Index : {node['law_index']}\nNode ID: {node.element_id}\nSimilarity: {similarity}\nNode Text: {node['text']}\n")
````
