````python
import os
from dotenv import load_dotenv
from openai import OpenAI


load_dotenv(verbose=True)

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")

client = OpenAI(
    api_key=OPENAI_API_KEY
)

PROMPT = """
당신은 HTML 문서를 Markdown 형식으로 변환하는 AI 어시스턴트입니다. HTML 태그를 파악하고, 이를 구조화된 텍스트 형태로 나타내어 사용자가 이해하기 쉽게 변환하는 것이 주요 임무입니다.

[작업 지시사항]
- <img>의 경우 화살표와 같이 글의 구조를 나타내는 경우가 있으므로, <img> 또한 사람이 이해하기 쉽게 구조화하여 텍스트 형태로 나타내야 합니다.
- 이미지 태그(<img>)에서 화살표가 사용된 경우, 텍스트 기반의 화살표 기호(→)로 순서 및 흐름을 표현할 수 있습니다.
- <table>의 경우 markdown을 통해 변환해주세요. 이 때 HTML 본문에 나와있는 내용만을 사용해주세요.
- <code>, <pre>와 같은 코드 블록은 생략해주세요.
- <math> 또는 <equation>과 같은 수식 태그는 LaTeX 형식으로 변환하세요.
- 변환된 모든 텍스트는 UTF-8 인코딩 형식이어야 합니다.
- ```markdown``` 과 같은 형식 표시는 생략해주세요.

[요청사항]
위의 [작업 지시사항]을 바탕으로 HTML 문서를 적절하게 변환한 Markdown 문서를 반환하세요.
"""
````

````python
import os

ROOT_PATH = "../DCM/dart/"
RESULT_PATH = "../DCM/dart_GPT_markdown/"

file_paths = [
    os.path.join(ROOT_PATH, "240_합병/240_2_합병가액산정방법/합병가액산정방법.txt"),
    os.path.join(ROOT_PATH, "230_발행공시/230_2_업무절차/업무절차.txt"),
    os.path.join(ROOT_PATH, "310_대량보유상황보고/310_2_세부내역및유의사항등안내/세부내역및유의사항등안내.txt"),
    os.path.join(ROOT_PATH, "550_공시관련부서안내/공시관련부서안내.txt")
]

file_names = [os.path.basename(path) for path in file_paths]

# 파일 내용 리스트에 저장
file_contents = []

for file_path in file_paths:
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()
        file_contents.append(content)
````

````python
for idx in range(len(file_paths)):
    print(f"## file name : {file_names[idx]}")
    print(f"## file path : {file_paths[idx]}")
    print(f"## content length : {len(file_contents[idx])}")
    print("## content ##")

    print(file_contents[idx])
````

````python
from bs4 import BeautifulSoup

# 불필요한 공백 및 개행 제거 함수
def delete_newlines(html_text: str) -> str:
    # BeautifulSoup 객체 생성
    soup = BeautifulSoup(html_text, 'html.parser')

    for element in soup.find_all(text=True):
        if element.parent.name not in ['pre', 'textarea']:  # 원본 텍스트 보존 태그는 제외
            # 태그 내 공백을 유지한 채로 앞뒤 공백 및 개행 문자만 제거
            cleaned_text = element.strip()
            element.replace_with(cleaned_text)

    # soup를 다시 문자열로 변환하면서 불필요한 개행과 공백을 제거
    cleaned_html = ''.join(str(soup).splitlines())
    return cleaned_html
````

````python
delete_newline_contents = [delete_newlines(x) for x in file_contents]
````

````python
from GPT_cost_calculator import num_tokens_from_string, cost_calculator

print(num_tokens_from_string(delete_newline_contents[3], 'gpt-3.5-turbo'))
print(cost_calculator(len(delete_newline_contents[3]), 'gpt-3.5-turbo'))

print(num_tokens_from_string(file_contents[3], 'gpt-3.5-turbo'))
print(cost_calculator(len(file_contents[3]), 'gpt-3.5-turbo'))
````

````python
def get_gpt_answer(file_content, file_name, model_name):
    completion = client.chat.completions.create(
        model=model_name,
        messages=[
            {"role": "system", "content": PROMPT},
            {
                "role": "user",
                "content": file_content
            }
        ]
    )

    with open(os.path.join(RESULT_PATH, file_name), "w") as file:
        print(f"## {RESULT_PATH} 에 {file_name} 저장 ##")
        file.write(completion.choices[0].message.content)
    
    return completion.choices[0].message.content
````

## GPT-3.5-Turbo

````python
get_gpt_answer(file_contents[3], file_names[3], "gpt-3.5-turbo")
````

````python
get_gpt_answer(delete_newline_contents[3], "del_new_" + file_names[3], "gpt-3.5-turbo")
````
