```python
import json

# 파일을 열고 JSON 데이터를 파싱하여 불러오기
with open("../results/1-2/DCM_1-2_law_main.json", "r", encoding="utf-8") as file:
    DCM_1_2_law_main = json.load(file)
```

```python
contents = []

for row in DCM_1_2_law_main:
    if type(row['content']) == str:
        contents.append(row['content'])
    elif type(row['content']) == list:
        contents += row['content']
    else:
        raise ValueError("not str or list")

len(contents)
```

```python
from bertopic import BERTopic
from konlpy.tag import Okt
from sklearn.feature_extraction.text import CountVectorizer

class CustomTokenizer:
    def __init__(self, tagger):
        self.tagger = tagger
    def __call__(self, sent):
        sent = sent[:1000000]
        word_tokens = self.tagger.morphs(sent)
        result = [word for word in word_tokens if len(word) > 1]
        return result

custom_tokenizer = CustomTokenizer(Okt())

vectorizer = CountVectorizer(tokenizer=custom_tokenizer, max_features=3000)

model = BERTopic(language="korean",
                 embedding_model="snunlp/KR-SBERT-V40K-klueNLI-augSTS", \
                 vectorizer_model=vectorizer,
                 nr_topics=30,
                 top_n_words=5,
                 verbose=True,
                 calculate_probabilities=True)

topics, probs = model.fit_transform(contents)
```
